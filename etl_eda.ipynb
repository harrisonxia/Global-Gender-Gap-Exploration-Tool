```python
# helper func
def get_indicator(df, num):
    return df.loc[df['indicator_id'] == num]

def get_nan_count(df, axis = 0):
    # default axis = 0 for overall
    # axis = 1 to get row count
    return df.isnull().sum(axis = axis)
```

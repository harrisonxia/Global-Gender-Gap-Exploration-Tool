```python
# helper func
def get_indicator(df, num):
    return df.loc[df['indicator_id'] == num]

def get_nan_count(df, axis = 0):
    # default axis = 0 for overall
    # axis = 1 to get row count
    return df.isnull().sum(axis = axis)
```

```python
def toJson(df, path):
    years = list(range(1990, 2018))
    years.append(9999)

    json_data = []
    for (index, row) in df.fillna(0).iterrows():
        row_obj = {
        'label': row['iso3'],
        }
        yearly = []
        for year in years:
             yearly.append(row[year])
        row_obj['data'] = yearly
        json_data.append(row_obj)
    
    json_obj = {
        'year': years,
    }
    json_obj['data'] = json_data
#     print(json_obj)
    with open(path, 'w') as fp:
        json.dump(json_obj, fp)


toJson(life_expectancy_female, './data/life_expectancy_female.json')
```
